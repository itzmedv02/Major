```python
data=pd.read_csv("rainfall.csv")
```

```python
import numpy as np  # Make sure numpy is imported

print("Data heads:")
print(data.head())

print("Null values in the dataset before preprocessing:")
print(data.isnull().sum())

print("Filling null values with mean of that particular column")
data = data.fillna(data.select_dtypes(include=[np.number]).mean())

print("Mean of numeric columns in data:")
print(data.select_dtypes(include=[np.number]).mean())

print("Null values in the dataset after preprocessing:")
print(data.isnull().sum())

print("\n\nShape: ", data.shape)
```

```python
print("Info:")
print(data.info())
```

```python
print("Group by:")
data.groupby('SUBDIVISION').size()
```

```python
print("Co-Variance =")
print(data.select_dtypes(include=[np.number]).cov())

print("\nCo-Relation =")
print(data.select_dtypes(include=[np.number]).corr())
```

```python
# Select only numeric columns for correlation calculation
numeric_data = data.select_dtypes(include=[np.number])
corr_cols = numeric_data.corr()['ANNUAL'].sort_values(ascending=False)
print("Index of correlation columns:", corr_cols.index)
```
